### Define Fee Revenue Plot in Code

```python
def Fee_revenue_plot(experiments,test_title,T):

    """
    This function plots the revenues generated for different assets.
    Each for a different static fee percentage (1%, 2% & 5%)
    """
    revenue_for_i_1 = []
    revenue_for_i_2 = []
    revenue_for_i_3 = []
    revenue_for_j_1 = [] 
    revenue_for_j_2 = []
    revenue_for_j_3 = []
    timestep_arr = []
    
    #Define timesteps as seen in plot_utils.py
    df = experiments
    df = df[df['substep'] == df.substep.max()]
    df.fillna(0,inplace=True)
    df.reset_index()

    #Append i & j revenue values to corresponding array 
    for i in range(df.substep.max(), T, df.substep.max()):
        revenue_for_i_1.append(df.iloc[i]['fee_revenue']['i'])
        revenue_for_j_1.append(df.iloc[i]['fee_revenue']['j'])
        timestep_arr.append(df.iloc[i]['timestep'])
        
        revenue_for_i_2.append(df.iloc[i+300]['fee_revenue']['i'])
        revenue_for_j_2.append(df.iloc[i+300]['fee_revenue']['j'])
        
        revenue_for_i_3.append(df.iloc[i+600]['fee_revenue']['i'])
        revenue_for_j_3.append(df.iloc[i+600]['fee_revenue']['j'])
        
    #Plot each revenue, each with unique asset (i or j) and percent (1%, 2% & 5%)
    plt.plot(timestep_arr, revenue_for_i_1, label='i revenue for 1%')
    plt.plot(timestep_arr, revenue_for_j_1, label='j revenue for 1%')
    plt.plot(timestep_arr, revenue_for_i_2, label='i revenue for 2%')
    plt.plot(timestep_arr, revenue_for_j_2, label='j revenue for 2%')
    plt.plot(timestep_arr, revenue_for_i_3, label='i revenue for 5%')
    plt.plot(timestep_arr, revenue_for_j_3, label='j revenue for 5%')
    plt.xlabel('Timestep')
    plt.ylabel('Arbitrage Wedge')
    plt.title("Fee Revenue for different fee percentages: 1%, 2% & 5% (oracle i = 1)")
    plt.legend()
    plt.show()
```

```python
Fee_revenue_plot(rdf,'Arbitrage Wedge Omnipool', 100)
```
